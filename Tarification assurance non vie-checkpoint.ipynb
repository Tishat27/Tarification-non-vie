Celle de la sévérité dénommée **d_sev**:

IDpol : Identifiant de la police d’assurance (utilisé pour faire le lien avec le jeu de données des sinistres).

ClaimAmount : Montant d'un sinistre


```python
#Installation des packages de base de R
utils = importr('utils')
#base = importr('base')

utils.chooseCRANmirror(ind=1)
utils.install_packages('CASdatasets')

cas=importr('CASdatasets')

robjects.r('data(freMTPL2freq)')
robjects.r('data(freMTPL2sev)')

# Récupérer l’objet freMTPL

df_freq= robjects.r('freMTPL2freq')
df_sev=robjects.r('freMTPL2sev')

# Le convertir en DataFrame pandas

df = pandas2ri.rpy2py(df_freq)
d_sev=pandas2ri.rpy2py(df_sev)
```

Certaines vérifications sont effectuées sur les deux bases avant de les joindre. Celles ci concernent principalement le fait que les polices de **d_sev** se retrouvent effectivement dans la base contenant les caractéristiques des polices **df**
Il n'y a pas de doublons de police dans **df**.


Les résultats présentent que les polices dans **df** sont uniques cependant dans il existe 6 polices avec des montants de sinistre qui n'existe pas dans la base **df**. Par conséquent elles ont été rétirées de l'analyse. Nous avons également vérifiées que leur retrait n'avait pas un impact significatif sur le volume des réclamations qui finit à **98%** après leur retrait.

```python
# Vérification des polices uniques et communes
results = []

unique_policies = df['IDpol'].nunique()
results.append(["Polices uniques dans df", unique_policies])

common_policies = set(d_sev['IDpol']).intersection(set(df['IDpol']))
results.append(["Polices communes entre d_sev et df", len(common_policies)])

# Création du DataFrame pour afficher les résultats sous forme de table
df_results = pd.DataFrame(results, columns=["Vérification", "Valeur"])
print(df_results)

##6 polices n'ont pas de caractéristiques dans la table principale
TotalClaim_b=d_sev['ClaimAmount'].sum()
non_common_policies = d_sev[~d_sev['IDpol'].isin(df['IDpol'])]
d_sev = d_sev[d_sev['IDpol'].isin(df['IDpol'])]
##Multiples lignes pour une police dans sev
sev=d_sev.groupby("IDpol").agg(ClaimAmount=("ClaimAmount","sum"),
                         Claimsevnb=("IDpol","count")).reset_index()

TotalClaim_a=d_sev['ClaimAmount'].sum()

TotalClaim_a/TotalClaim_b
```

```python
##Joindre les deux bases
data= pd.merge(df,sev,on="IDpol",how="outer")
data['ClaimAmount'].fillna(0, inplace = True)
data['Claimsevnb'].fillna(0, inplace = True)
```

```python
#Étape 0 : Analyse exploratoire des données
###Vérification des données
print(data.info()) # Vérification des types de données
print(data.describe()) # Statistiques descriptives
len(data["IDpol"].unique())

# Vérification des plages de valeurs
plt.figure(figsize=(10,6))
sns.heatmap(data.isnull(), cmap="coolwarm", cbar=False)
plt.show()
data.isnull().sum()
```

```python
#Corrélation entre les variables
corr_matrix = data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

La corrélation entre les variables joue un rôle déterminant dans la performance et la stabilité d'un modèle. En présence d'une forte corrélation, plusieurs problèmes peuvent survenir :

 -La matrice des coefficients peut devenir singulière, ce qui signifie qu'elle n'est pas inversible.
 -Certaines valeurs propres (eigenvalues) peuvent être nulles ou proches de zéro, indiquant une redondance dans les données.
 -Si l'on tente d'inverser une telle matrice, on risque d'obtenir une variance très élevée, entraînant une instabilité numérique et de larges erreurs dans les prédictions.
 -Il devient alors difficile, voire impossible, de déterminer l'impact réel des variables corrélées sur le modèle.

Par précaution, il est recommandé de supprimer l'une des variables fortement corrélées, afin d'améliorer la robustesse et l'interprétabilité du modèle.
